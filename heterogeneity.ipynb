```python
mat = pd.read_table('/home/xieyufeng/ex/output/scirep/matrix_processing/filter.scimpute_count.Norm_SCnorm.Batch_Combat_1.domains_combined.txt',index_col=0)
scirep = np.log2(mat.iloc[np.where(np.isin([i.split('|')[0] for i in mat.index],ref))]+1)
stack = pd.DataFrame(scirep.stack())
stack = stack.reset_index()
stack.columns=['feature','sample_id','log2(count+1)']
merge = pd.merge(stack,class_info,on=['sample_id'])
plt.subplots(figsize=(15,15))
ax = sns.stripplot(x="label", y="log2(count+1)",hue='feature', data=merge,jitter=0.4)
```

```python
mat_raw = pd.read_table('/home/xieyufeng/ex/output/scirep/matrix_processing/filter.scimpute_count.Norm_SCnorm.Batch_Combat_1.domains_combined.txt',index_col=0)
scirep_raw = np.log2(mat_raw.iloc[np.where(np.isin([i.split('|')[0] for i in mat_raw.index],ref))]+1)
stack_raw = pd.DataFrame(scirep_raw.stack())
stack_raw = stack_raw.reset_index()
stack_raw.columns=['feature','sample_id','log2(count+1)']
merge_raw = pd.merge(stack_raw,class_info,on=['sample_id'])
merge_raw['precess']='normalized'
fig,(ax,ax1) = plt.subplots(1,2,figsize=(30,15))
sns.stripplot(ax=ax,x="label", y="log2(count+1)",hue='feature', data=merge_raw,jitter=0.4)

mat = pd.read_table('/home/xieyufeng/ex/output/scirep/count_matrix/domains_combined.txt',index_col=0)
scirep = np.log2(mat.loc[scirep_raw.index]+1)
stack = pd.DataFrame(scirep.stack())
stack = stack.reset_index()
stack.columns=['feature','sample_id','log2(count+1)']
merge = pd.merge(stack,class_info,on=['sample_id'])
merge['precess']='unnormalized'
sns.stripplot(ax=ax1,x="label", y="log2(count+1)",hue='feature', data=merge,jitter=0.4)
```
